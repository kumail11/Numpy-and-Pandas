# Missing Data
### isnull()
### notnull()
### fillna()

```python
data.isnull()
```

```python
pd.notnull(data)
```

```python
data['Age'].isnull()
```

```python
data['Age'].isnull().value_counts()
```

```python
data['Age'].fillna(20,inplace=True)
data
```

```python
data['Age'].value_counts()
```

```python
#Filling null values with the previous ones
data.fillna(method='pad')
```

```python
# Filling null value with the next ones
data.fillna(method='bfill')
```

```python
# Filling a null values using replace() method
data['CompTotal'].replace(to_replace = np.nan, value = 50) 
```

```python
#Dropping rows if all values in that row are missing.
data.dropna
```

```python
data.dropna(how = 'all') 
```

```python
# Dropping columns with at least 1 null value.
data.dropna(axis=1, how='all')
```

```python
#Dropping Rows with at least 1 null value in CSV file
data.shape
data.dropna(axis=0, how='any', inplace =True)
data.shape
```
