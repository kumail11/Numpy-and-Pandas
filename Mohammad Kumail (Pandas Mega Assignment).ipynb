```python
# 69): Write a Pandas program to merge datasets and check uniqueness.

print("Original Frame:")
print(select_by_datatype_frame)

copy_of_datatype_frame = select_by_datatype_frame.copy(deep=True)
temp_frame1 = select_by_datatype_frame.drop([0, 1])
copy_of_datatype_frame = copy_of_datatype_frame.drop([2])

print("---------------------------------------")

print("New Frame")
print(temp_frame1)
print(copy_of_datatype_frame)

print("---------------------------------------")

print('"one_to_one": check if merge keys are unique in both left and right datasets:"')
one_to_one = pd.merge(temp_frame1, copy_of_datatype_frame, validate="one_to_one")
print(one_to_one)

print("---------------------------------------")

print('"one_to_many" or "1:m": check if merge keys are unique in left dataset:')
one_to_many = pd.merge(temp_frame1, copy_of_datatype_frame, validate="one_to_many")
print(one_to_many)

print("---------------------------------------")

print('“many_to_one” or “m:1”: check if merge keys are unique in right dataset:')
many_to_one = pd.merge(temp_frame1, copy_of_datatype_frame, validate="many_to_one")
print(many_to_one)
```

```python
# 78): Write a Pandas program to replace the current value in a dataframe column based on last largest value.
#     If the current value is less than last largest value replaces the value with 0.


rand_dict = {'Random_Number':[23, 21, 27, 22, 34, 33, 34, 31, 25, 22, 34, 19, 31, 32, 19]}
rand_num_frame = pd.DataFrame(rand_dict)

print("Original Frame")
print(rand_num_frame)


print("------------------------------------------------------------------------------------------")

print("Replace current value in a dataframe column based on last largest value:")

rand_num_frame['Random_Number'] = rand_num_frame.Random_Number.where(rand_num_frame.Random_Number.eq(
    rand_num_frame.Random_Number.cummax()), 0)

print(rand_num_frame)
```

```python
# 80): Write a Pandas program to check for inequality of two given DataFrames. 

inequal_dict1 = {'W':[68,75,86,80,None],'X':[78,85,None,80,86], 'Y':[84,94,89,83,86],'Z':[86,97,96,72,83]}
inequal_frame1 = pd.DataFrame(inequal_dict1)
print(inequal_frame1)

print("---------------------------------------")

inequal_dict2 = {'W':[78,75,86,80,None],'X':[78,85,96,80,76], 'Y':[84,84,89,83,86],'Z':[86,97,96,72,83]}
inequal_frame2 = pd.DataFrame(inequal_dict2)
print(inequal_frame2)

print("---------------------------------------")

print("Check for inequality of the said dataframes:")
print(inequal_frame1.ne(inequal_frame2))
```

```python
# 81): Write a Pandas program to get lowest n records within each group of a given DataFrame.

smallest_dict = {'col1': [1, 2, 3, 4, 7, 11], 'col2': [4, 5, 6, 9, 5, 0], 'col3': [7, 5, 8, 12, 1,11]}

smallest_frame = pd.DataFrame(smallest_dict)
print(smallest_frame)

print("---------------------------------------")

print("Lowest n records within each group of a DataFrame:")

small_1 = smallest_frame.nsmallest(3, 'col1')
print(small_1)

print("---------------------------------------")

small_2 = smallest_frame.nsmallest(3, 'col2')
print(small_2)

print("---------------------------------------")

small_3 = smallest_frame.nsmallest(3, 'col3')
print(small_3)
```
