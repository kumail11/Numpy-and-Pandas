```python
import pandas as pd
import numpy as np
```

```python
# 1): Load up the dataset and set header label names to:

header = ['education', 'age', 'capital-gain', 'race', 'capital-loss', 'hours-per-week', 'sex', 'classification']

myframe = pd.read_csv('census.data', names = header)
myframe
```

```python
myframe.head(5)
```

```python
myframe.tail(5)
```

```python
# 2): Make sure any value that needs to be replaced with a NAN is set as such. There are at
# least three ways to do this. One is much easier than the other two.

myframe = myframe.replace({'?': np.nan, '&': np.nan, '+': np.nan})
myframe
```

```python
myframe = myframe.fillna(method='ffill')
myframe
```

```python
# 3): Look through the dataset and ensure all of your columns have appropriate data types.
# Numeric columns should be float64 or int64, and textual columns should be object.

myframe.dtypes
```

```python
myframe['capital-gain'] = pd.to_numeric(myframe['capital-gain'], errors='coerce')
```

```python
myframe.dtypes
```

```python
myframe.education.unique()
```

```python
# 4): Properly encode any ordinal features using the method discussed in the chapter.

ordered_education = ['1st-4th', '5th-6th', '7th-8th', '9th', '10th', '11th',
                     '12th', 'Bachelors', 'Masters']

myframe = pd.DataFrame({'education': ['Bachelors', '12th', 'Masters', '10th', '9th']})

myframe.education = myframe.education.astype("category", ordered=True, categories=ordered_education).cat.codes
```

```python
ordinal = pd.get_dummies(myframe, columns=['education'])
ordinal
```

```python
myframe.race.unique()
```

```python
# 5): Properly encode any nominal features by exploding them out into new, separate,
# boolean features.

unordered_gender = ['Male', 'Female']

myframe = pd.DataFrame({'sex': ['Female', 'Male', 'Male', 'Female', 'Female', 'Male']})
myframe.sex = myframe.sex.astype("category", ordered=True, categories=unordered_gender).cat.codes
```

```python
gender = pd.get_dummies(myframe, columns=['sex'])
gender
```

```python
unordered_race = ['White', 'Amer-Indian-Eskimo', 'Black', 'Asian-Pac-Islander']

myframe = pd.DataFrame({'race': ['Black', 'Asian-Pac-Islander', 'Black', 'White', 'Amer-Indian-Eskimo', 'White']})
myframe.race = myframe.race.astype("category", ordered=True, categories=unordered_race).cat.codes
```

```python
race = pd.get_dummies(myframe, columns=['race'])
race
```

```python
myframe.to_csv('newCensus.data')
myframe
```
